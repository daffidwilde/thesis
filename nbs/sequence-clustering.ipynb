```python
import json
from collections import defaultdict

import pandas as pd
```

```python
with open("/Volumes/thesis-data/dtype_dict.json") as f:
    dtypes = json.load(f)

main = pd.read_csv(
    "/Volumes/thesis-data/main.csv",
    dtype=dtypes,
    parse_dates=["PERIOD", "ADMDATE", "DISCDATE"],
)
```

```python
process_columns = []
for column in main.columns:
    if column.startswith("OPCS_"):
        try:
            int(column.split("_")[-1])
            process_columns.append(column)
        except ValueError:
            pass

process_columns = sorted(process_columns, key=lambda opcs: int(opcs.split("_")[-1]))

process_columns.insert(0, "OPCS4")
```

```python
primary_process = main[["SPELL_ID", "EPISODE_ID", "NetCost"] + process_columns]
primary_process = primary_process.drop_duplicates(subset=["EPISODE_ID"]).reset_index(
    drop=True
)

episode_counts = primary_process.groupby("SPELL_ID")["EPISODE_ID"].count()
multiple_episode_spells = episode_counts[episode_counts > 1].index

primary_process = primary_process[
    primary_process["SPELL_ID"].isin(multiple_episode_spells)
].reset_index(drop=True)
primary_process = primary_process.sort_values("EPISODE_ID")
```
